```python
import pandas as pd

dataset1 = pd.read_csv('data.tsv', delimiter='\t', encoding='utf-8')
dataset2 = pd.read_csv('data-2.tsv', delimiter='\t', encoding='utf-8')
dataset3 = pd.read_csv('data-3.tsv', delimiter='\t', encoding='utf-8')
dataset4 = pd.read_csv('data-4.tsv', delimiter='\t', encoding='utf-8')
dataset5 = pd.read_csv('data-5.tsv', delimiter='\t', encoding='utf-8')
dataset6 = dataset5[(dataset5["category"] == "actor") | (dataset5["category"] == "actress") | (dataset5["category"] == "director")]
```

```python
# Assuming dataset1, dataset2, dataset3, dataset4, and dataset6 are your DataFrames

# First, merge dataset1 and dataset2
merged_1_2 = pd.merge(dataset1, dataset2, on='tconst', how='inner')

# Then merge the result with dataset3
merged_1_2_3 = pd.merge(merged_1_2, dataset3, on='tconst', how='left')

# Continue the process with dataset4
merged_1_2_3_6 = pd.merge(merged_1_2_3, dataset6, on='tconst', how='left')

# Finally, merge with dataset6
merged_dataset = pd.merge(merged_1_2_3_6, dataset4, on='nconst', how='inner')

# Now, merged_dataset contains the combined data from all datasets
```

```python
movies_dataset= merged_dataset[merged_dataset ['titleType']=='movie']
```

```python
movies_dataset = movies_dataset.drop(columns=['runtimeMinutes', 'tconst', 'titleType', 'primaryTitle',  'originalTitle', 'isAdult', 'startYear', 'endYear', 'writers', 'ordering', 'nconst', 'job', 'characters', 'birthYear', 'deathYear', 'primaryProfession', 'knownForTitles'])
```

```python
movies_dataset.head()
```

```python
movies_dataset['genres'] = movies_dataset['genres'].str.split(',')

# Explode the 'genres' column into separate rows
exploded_genres = movies_dataset.explode('genres')

# Now, exploded_genres DataFrame has separate rows for each genre of a movie
```
