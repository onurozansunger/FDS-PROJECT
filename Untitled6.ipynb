```python
netflix4 = pd.read_csv('netflix4.csv')

# Gereksiz sütunları kaldırma (örnek olarak)

# Veri tipi optimizasyonu (örnek olarak)

# 'genres' ve 'primaryName' sütunlarını liste haline getirme
netflix4['genres'] = netflix4['genres'].apply(lambda x: x.split(',') if isinstance(x, str) else [])
netflix4['primaryName'] = netflix4['primaryName'].apply(lambda x: x.split(',') if isinstance(x, str) else [])

# One-hot encoding uygulama
mlb_genres = MultiLabelBinarizer(sparse_output=True)
genres_encoded = mlb_genres.fit_transform(netflix4['genres'])
genres_encoded_df = pd.DataFrame(genres_encoded.toarray(), columns=mlb_genres.classes_)

mlb_names = MultiLabelBinarizer(sparse_output=True)
names_encoded = mlb_names.fit_transform(netflix4['primaryName'])
names_encoded_df = pd.DataFrame(names_encoded.toarray(), columns=mlb_names.classes_)

# Kategoriye göre işleme
def categorize_names(names):
    categories = []
    for name in names:
        categories.append('director')  # Bu, örnek bir kategoridir.
    return categories

netflix4['name_categories'] = netflix4['primaryName'].apply(categorize_names)
mlb_categories = MultiLabelBinarizer(sparse_output=True)
categories_encoded = mlb_categories.fit_transform(netflix4['name_categories'])
categories_encoded_df = pd.DataFrame(categories_encoded.toarray(), columns=mlb_categories.classes_)

# DataFrame'lerin indexlerini sıfırlama ve birleştirme
netflix4.reset_index(drop=True, inplace=True)
genres_encoded_df.reset_index(drop=True, inplace=True)
names_encoded_df.reset_index(drop=True, inplace=True)
categories_encoded_df.reset_index(drop=True, inplace=True)

df_encoded = pd.concat([netflix4, genres_encoded_df, names_encoded_df, categories_encoded_df], axis=1)

# Gereksiz sütunları kaldırma
df_encoded.drop(['genres', 'primaryName', 'name_categories'], axis=1, inplace=True)

# Yeni DataFrame'in ilk beş satırını gösterme
print(df_encoded.head())
```
