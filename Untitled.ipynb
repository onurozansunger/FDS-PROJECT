```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MultiLabelBinarizer
import numpy as np
```

```python
dataset1 = pd.read_csv('data.tsv', delimiter='\t', encoding='utf-8')
dataset2 = pd.read_csv('data-2.tsv', delimiter='\t', encoding='utf-8')
dataset3 = pd.read_csv('data-3.tsv', delimiter='\t', encoding='utf-8')
dataset4 = pd.read_csv('data-4.tsv', delimiter='\t', encoding='utf-8')
dataset5 = pd.read_csv('data-5.tsv', delimiter='\t', encoding='utf-8')
dataset6 = dataset5[(dataset5["category"] == "actor") | (dataset5["category"] == "actress") | (dataset5["category"] == "director")]
netflix = pd.read_csv('netflix_titles.csv')
```

```python
# First, merge dataset1 and dataset2
merged_1_2 = pd.merge(dataset1, dataset2, on='tconst', how='inner')

# Then merge the result with dataset3
merged_1_2_3 = pd.merge(merged_1_2, dataset3, on='tconst', how='left')

# Continue the process with dataset4
merged_1_2_3_6 = pd.merge(merged_1_2_3, dataset6, on='tconst', how='left')

# Finally, merge with dataset6
merged_dataset = pd.merge(merged_1_2_3_6, dataset4, on='nconst', how='inner')
```

```python
netflix4['genres'] = netflix4['genres'].str.split(',')

# Explode the 'genres' column into separate rows
netflix5 = netflix4.explode('genres')

# Now, exploded_genres DataFrame has separate rows for each genre of a movie
```

```python

# 'genres' sütununu liste haline getirme ve one-hot encoding uygulama
netflix5['genres'] = netflix5['genres'].apply(lambda x: x.split(',') if isinstance(x, str) else [])
mlb_genres = MultiLabelBinarizer()
genres_encoded = mlb_genres.fit_transform(netflix5['genres'])
genres_encoded_df = pd.DataFrame(genres_encoded, columns=mlb_genres.classes_)

# 'primaryName' sütununu işleme
netflix5['primaryName'] = netflix5['primaryName'].apply(lambda x: x.split(',') if isinstance(x, str) else [])
mlb_names = MultiLabelBinarizer()
names_encoded = mlb_names.fit_transform(netflix5['primaryName'])
names_encoded_df = pd.DataFrame(names_encoded, columns=mlb_names.classes_)

# Kategoriye göre işleme (örneğin, director, actor, actress)
def categorize_names(names):
    categories = []
    for name in names:
        # Burada, ismin kategorisini belirleyen bir mantık olmalı.
        # Örnek olarak, her ismi 'director' olarak işaretliyorum.
        categories.append('director')  # Gerçek uygulamada bu, veri setinizdeki gerçek verilere göre değişmelidir.
    return categories

netflix5['name_categories'] = netflix5['primaryName'].apply(categorize_names)
mlb_categories = MultiLabelBinarizer()
categories_encoded = mlb_categories.fit_transform(netflix5['name_categories'])
categories_encoded_df = pd.DataFrame(categories_encoded, columns=mlb_categories.classes_)

# Son olarak, orijinal DataFrame ile encoded DataFrame'leri birleştirme
# ignore_index=True parametresini kaldırıyorum çünkü sütun isimlerini korumak istiyoruz.
df_encoded = pd.concat([netflix5, genres_encoded_df, names_encoded_df, categories_encoded_df], axis=1)

# Gereksiz sütunları kaldırma
df_encoded.drop(['genres', 'primaryName', 'name_categories'], axis=1, inplace=True)

# Yeni DataFrame'in ilk beş satırını gösterme
print(df_encoded.head())
```

```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer

# 'genres' sütununu liste haline getirme ve one-hot encoding uygulama
netflix4['genres'] = netflix4['genres'].apply(lambda x: x.split(',') if isinstance(x, str) else [])
mlb_genres = MultiLabelBinarizer()
genres_encoded = mlb_genres.fit_transform(netflix4['genres'])
genres_encoded_df = pd.DataFrame(genres_encoded, columns=mlb_genres.classes_)

# 'primaryName' sütununu işleme
netflix4['primaryName'] = netflix4['primaryName'].apply(lambda x: x.split(',') if isinstance(x, str) else [])
mlb_names = MultiLabelBinarizer()
names_encoded = mlb_names.fit_transform(netflix4['primaryName'])
names_encoded_df = pd.DataFrame(names_encoded, columns=mlb_names.classes_)

# Kategoriye göre işleme
def categorize_names(names):
    categories = []
    for name in names:
        categories.append('director')  # Bu, örnek bir kategoridir.
    return categories

netflix4['name_categories'] = netflix4['primaryName'].apply(categorize_names)
mlb_categories = MultiLabelBinarizer()
categories_encoded = mlb_categories.fit_transform(netflix4['name_categories'])
categories_encoded_df = pd.DataFrame(categories_encoded, columns=mlb_categories.classes_)

# DataFrame'lerin indexlerini sıfırlama
netflix4.reset_index(drop=True, inplace=True)
genres_encoded_df.reset_index(drop=True, inplace=True)
names_encoded_df.reset_index(drop=True, inplace=True)
categories_encoded_df.reset_index(drop=True, inplace=True)

# DataFrame'leri birleştirme
df_encoded = pd.concat([netflix4, genres_encoded_df, names_encoded_df, categories_encoded_df], axis=1)

# Gereksiz sütunları kaldırma
df_encoded.drop(['genres', 'primaryName', 'name_categories'], axis=1, inplace=True)

# Yeni DataFrame'in ilk beş satırını gösterme
print(genres_encoded_df.head())
```
